```
def inception(x,nb_filter):
    branch1x1 = Conv2D(nb_filter, (1,1), padding='same')(x)
    branch3x3 = Conv2D(nb_filter, (1,1), padding='same')(x)
    branch3x3 = Conv2D(nb_filter,(3,3), padding='same')(branch3x3)
 
    branch5x5 = Conv2D(nb_filter,(1,1), padding='same')(x)
    branch5x5 = Conv2D(nb_filter,(5,5), padding='same')(branch5x5)
 
    branchpool = MaxPooling2D(pool_size=(3,3), strides=(1,1), padding='same')(x)
    branchpool = Conv2D(nb_filter, (1,1),padding='same')(branchpool)
 
    x = concatenate([branch1x1,branch3x3,branch5x5,branchpool],axis=3)
 
    return x
```
